```python
import pandas as pd
```

```python
df = pd.read_csv('Documents/intern_data.csv')
```

```python
df.rename(columns={'Total Revenue (M)': 'Total Revenue'}, inplace=True)
df.rename(columns={'Net Income (M)': 'Net Income '}, inplace=True)
df.rename(columns={'Total Assets (M)': 'Total Assets'}, inplace=True)
df.rename(columns={'Total Liabilities (M)': 'Total Liabilities '}, inplace=True)
```

```python
print(df.columns)
```

```python

df.columns = df.columns.str.strip()


df['Revenue Growth (%)'] = df['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df['Net Income'].pct_change() * 100
```


Review of the Analysis

In this section, we have calculated the financial growth for each company (Microsoft, Tesla, Apple) over the past few years. We have derived the revenue growth, net income growth, and other key financial metrics. Letâ€™s confirm the calculations:

1. **Revenue Growth**: We computed the percentage change in total revenue year-over-year for each company.
2. **Net Income Growth**: Similarly, we computed the percentage change in net income.
3. **Other financial metrics**: We also reviewed other relevant financial metrics to check consistency and trends.

Now, let's visualize these results for better understanding.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.lineplot(data=df, x='Year', y='Revenue Growth (%)', hue='Company')
plt.title('Revenue Growth Over Years by Company')
plt.xlabel('Year')
plt.ylabel('Revenue Growth (%)')
plt.show()
```
